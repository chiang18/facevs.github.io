```python
# load and evaluate a saved model
import os
from numpy import loadtxt
from keras.models import load_model
from keras.preprocessing import image
import numpy as np
```

```python
image_dir = '../image'
model_dir = '../model'
predict_dir = "./pre"
```

```python
for mfile in os.listdir(image_dir):
    print(mfile)
```

```python
def predict_prob(number):
  return [number[0],1-number[0]]
```

```python
# dimensions of our images
img_width, img_height = 300, 300


# load the model we saved
Model_dir = model_dir + "/"
for mfile in os.listdir(model_dir):
  model = load_model(Model_dir + str(mfile))
  model.compile(loss='binary_crossentropy',
                optimizer='rmsprop',   #origin is rmsprop
                metrics=['accuracy'])

  with open(predict_dir + "/" + mfile.rstrip("h5") + "txt", "w") as wfile:
    # predicting images
    Image_dir = image_dir + "/"
    for ifile in os.listdir(image_dir):
      img = image.load_img(Image_dir + str(ifile), target_size=(img_width, img_height))
      x = image.img_to_array(img)
      x = x/255.
      x = np.expand_dims(x, axis=0)

      images = np.vstack([x])
      classes = np.array(list(map(predict_prob, model.predict(images))))

      wfile.write("{} {} {}\n".format(ifile, classes[0][0], classes[0][1]))   #Probability of (Real, Fake)
      #print(ifile, "finish\n")
      print(classes[0][0])
  print(mfile, "finish\n")
    
```
